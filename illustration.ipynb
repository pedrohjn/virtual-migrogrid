```python
#Import libraries
import matplotlib.pyplot as plt #https://matplotlib.org/
import numpy as np #http://www.numpy.org/
```

```python
#t = ElectricityConsumption['2018-3-31'].index
t = np.arange(4*24)
x1= np.random.poisson(4,4*24)
x2=np.random.poisson(1.5,4*24)
x4=np.random.poisson(0.5,4*24)
x3=np.zeros(4*24)
```

```python
#One day
# plot it!
fig, ax = plt.subplots(1,figsize=(16,6))
ax.step(t, x1, lw=2, label='day 1', color='blue',linestyle='-', where='post')
ax.step(t, x1 + x2, lw=2, label='day 2', color='black', where='post')
ax.step(t, x1 + x2 + x4, lw=2, label='day 2', color='black', where='post')
ax.fill_between(t, x1 + x2, x1, facecolor='blue',step="post",  alpha=0.5, hatch="+")
ax.fill_between(t, x3, x1, facecolor='yellow',step="post",  alpha=0.5)
ax.fill_between(t, x1 + x2 + x4, x1 + x2 , facecolor='white', step="post",  alpha=0.5, hatch="//")
fig, ax = plt.subplots(1,figsize=(16,6))ax.grid()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patheffects as PathEffects
import numpy as np

fig, ax1 = plt.subplots(1,figsize=(16,6))
ax1.imshow([[1, 2], [2, 5]])
pe = [PathEffects.withStroke(linewidth=3,foreground="w")]
ax1.grid(True, linestyle="-", path_effects=pe)

plt.show()
```
